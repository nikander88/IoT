```python
#import libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates 
import numpy as np
```

```python
Temp1 = pd.read_csv("year.csv", dayfirst=True, sep=",",
header=0, decimal=b".",index_col=0, parse_dates= [[0, 1, 2, 3]], usecols=[0,1,2,3,5]) 
Temp1.info()

Temp2 = pd.read_csv("month.csv", dayfirst=True, sep=",",
header=0, decimal=b".",index_col=0, parse_dates= [[0, 1, 2, 3]], usecols=[0,1,2,3,5]) 
Temp2.info()

Temp3 = pd.read_csv("week.csv", dayfirst=True, sep=",",
header=0, decimal=b".",index_col=0, parse_dates= [[0, 1, 2, 3]], usecols=[0,1,2,3,5]) 
Temp3.info()

Temp4 = pd.read_csv("day.csv", dayfirst=True, sep=",",
header=0, decimal=b".",index_col=0, parse_dates= [[0, 1, 2, 3]], usecols=[0,1,2,3,5]) 
Temp4.info()
```

```python
plt.figure(figsize=(20,10)) 
plt.plot(Temp1,color='blue', marker='.',linestyle='-') 
plt.title("Air temperature in aug 2018 to aug 2019, at Lappeenranta airport station") 
plt.ylabel("Temperature in [C]") 
plt.grid(True) 
plt.show()
```

```python
plt.figure(figsize=(20,10)) 
plt.plot(Temp2,color='blue', marker='.',linestyle='-') 
plt.title("Air temperature in august, 2020, at Lappeenranta airport station") 
plt.ylabel("Temperature in [C]") 
plt.grid(True) 
plt.show()
```

```python
plt.figure(figsize=(20,10)) 
plt.plot(Temp3,color='blue', marker='.',linestyle='-') 
plt.title("Air temperature in aug-sep week in 2020, at Lappeenranta airport station") 
plt.ylabel("Temperature in [C]") 
plt.grid(True) 
plt.show()
```

```python
plt.figure(figsize=(20,10)) 
plt.plot(Temp4,color='blue', marker='.',linestyle='-') 
plt.title("Air temperature in 10th of september, at Lappeenranta airport station") 
plt.ylabel("Temperature in [C]") 
plt.grid(True) 
plt.show()
```
